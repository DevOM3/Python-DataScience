Plotting 2 Graphs

```python
from matplotlib import pyplot as plt

plt.title("MyData")
plt.xlabel("X Side")
plt.ylabel("Y Side")
x = [31, 23, 75]
y = [23, 74, 36]
x1 = [11, 22, 33]
y1 = [55, 66, 88]

plt.plot(x, y)
plt.plot(x1, y1)
plt.show()
```

## Changing Plot Style


```python
from matplotlib import style
from matplotlib import pyplot as plt

style.use('ggplot')
plt.title("My Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot([11, 21, 31, 14], [22, 33, 1, 89])
plt.show()
```

# Showing Legend

```python
from matplotlib import pyplot as plt

plt.title("MyData")
plt.xlabel("X Side")
plt.ylabel("Y Side")
x = [31, 23, 75]
y = [23, 74, 36]
x1 = [11, 22, 33]
y1 = [55, 66, 88]

plt.plot(x, y, label="first")
plt.plot(x1, y1, label="second")
plt.legend()
plt.show()
```
